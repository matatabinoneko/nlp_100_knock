## 70. データの入手・整形
[文に関する極性分析の正解データ](http://www.cs.cornell.edu/people/pabo/movie-review-data/rt-polaritydata.tar.gz)を用い，以下の要領で正解データ（sentiment.txt）を作成せよ．

rt-polarity.posの各行の先頭に"+1 "という文字列を追加する（極性ラベル"+1"とスペースに続けて肯定的な文の内容が続く）
rt-polarity.negの各行の先頭に"-1 "という文字列を追加する（極性ラベル"-1"とスペースに続けて否定的な文の内容が続く）
上述1と2の内容を結合（concatenate）し，行をランダムに並び替える
sentiment.txtを作成したら，正例（肯定的な文）の数と負例（否定的な文）の数を確認せよ．

```python
import random


tmp= []
with open("rt-polaritydata/rt-polarity.pos",'r',encoding='latin-1') as fin:
    line = fin.readline().strip('\n').strip(' ')
    while line:
        tmp.append("+1 " + line)
        line = fin.readline().strip('\n').strip(' ')

with open("rt-polaritydata/rt-polarity.neg",'r',encoding='latin-1') as fin:
    line = fin.readline().strip('\n').strip(' ')
    while line:
        tmp.append("-1 " + line)
        line = fin.readline().strip('\n').strip(' ')

random.shuffle(tmp)


with open("sentiment.txt",'w') as fout:
    fout.writelines('\n'.join(tmp))
    
```

## 71. ストップワード
英語のストップワードのリスト（ストップリスト）を適当に作成せよ．さらに，引数に与えられた単語（文字列）がストップリストに含まれている場合は真，それ以外は偽を返す関数を実装せよ．さらに，その関数に対するテストを記述せよ．

```python
stop_word = []
with open("stopwords.txt",'r') as f:
    line = f.readline().strip()
    while line:
        stop_word.append(line)
        line = f.readline().strip()

    
```

```python
def stopList(sentence):
    sentence = sentence.split(' ')
    if 0 < len(set(stop_word) & set(sentence)):
        return True
    else:
        return False
```

## 72. 素性抽出
極性分析に有用そうな素性を各自で設計し，学習データから素性を抽出せよ．素性としては，レビューからストップワードを除去し，各単語をステミング処理したものが最低限のベースラインとなるであろう

```python
word_list = []
with open("positive-words.txt",'r') as p,  open("negative-words.txt",'r') as n:
    line = p.readline().strip()
    while line:
        word_list.append(line)
        line = p.readline().strip()
    line = n.readline().strip()
    while line:
        word_list.append(line)
        line = n.readline().strip()
```

```python
print(word_list)
```

```python
from stemming.porter2 import stem

word_list = [stem(word) for word in word_list]

def makeSosei(sentence_list):
    ret = []
    for sentence in sentence_list:
        sentence = [word.lower() for word in sentence.split(" ")]
        sentence = list(set(sentence) - set(stop_word))
        sentence = [stem(word) for word in sentence]

        sosei = [0 for i in range(len(word_list))]
        for i,word in enumerate(word_list):
            sosei[i] = 1 if word in sentence else 0
        ret.append(sosei)
    return ret

```

```python
import numpy as np
with open("sentiment.txt",'r') as f:
    s_l = f.readlines()
    label_list = list(map(int,[l[:2]for l in s_l]))
    sentence_list = [s[3:] for s in s_l]
    feat_list = makeSosei(sentence_list)
    
```

## 73. 学習
72で抽出した素性を用いて，ロジスティック回帰モデルを学習せよ．

```python
import sklearn
from sklearn.linear_model import LogisticRegression

X,y = feat_list,label_list

scaler=sklearn.preprocessing.StandardScaler()
scaler.fit(X)
X=scaler.transform(X)

clf = LogisticRegression(random_state=0,solver='lbfgs').fit(X, y)
```

```python
sentence = ["I like it ."]
label = [1]
feat = makeSosei(sentence)
print("answer label is {}\npredict label is {}".format(label,clf.predict(feat)))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader 

class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(len(word_list),20)
        self.fc2 = nn.Linear(20,1)
        self.tanh = nn.Tanh()
        
    def forward(self,x):
        x = torch.tensor(x).float()
        x = self.fc1(x)
        x = self.fc2(F.relu(x))
        x = self.tanh(x)
        return x
    
        
```

```python
batch_size = 64
dataset = torch.utils.data.TensorDataset(torch.FloatTensor(feat_list),torch.FloatTensor(label_list))
data_loader = DataLoader(dataset,batch_size=batch_size,shuffle=True)
```

```python
class Model():
    def __init__(self):
        self.net = Net()
        self.n_epoch = 100
        self.optimizer = optim.Adam(self.net.parameters())
        
    def trainning(self,data_loader):
        self.net.train()
        for epoch in range(self.n_epoch):
            for batch in data_loader:
                x,t = batch
                y = self.net(x)
                loss = torch.mean((-1/2)*((1-t)*torch.log(abs(1-y)) + (1+t)*torch.log(abs(1+y))))
                self.optimizer.zero_grad()
                loss.backward()
                self.optimizer.step()
                
    def evaluation(self,feat,label):
        label = torch.tensor(label).long()
        self.net.eval()
        with torch.no_grad():
            y = self.net(feat)
            pred_label = torch.sign(y).long()
#             pred_label = 1 if 0 < torch.squeeze(y) else -1
        for i in range(len(label)):
            print("correst label is {}\npredict label is {}\nratio is {:<.2f}".format(torch.squeeze(label[i]).item(),torch.squeeze(pred_label[i]).item(),torch.squeeze(y[i]).item()))
            
    def cal(self,feat,label):
        self.net.eval()
        with torch.no_grad():
            y = self.net(feat)
            y = torch.squeeze(y)
            pred_label = torch.sign(y).long()
        return tuple(zip(label, pred_label.numpy().tolist(),y.numpy().tolist()))
```

```python
model = Model()
```

```python
model.trainning(data_loader)
```

## 76. ラベル付け
学習データに対してロジスティック回帰モデルを適用し，正解のラベル，予測されたラベル，予測確率をタブ区切り形式で出力せよ．

```python
list_77 = []
with torch.no_grad():
    model.net.eval()
    y = model.net(feat_list).detach().numpy()

for data in zip(y,label_list):
    out , t= data
    out = out[0]
#     t = t.astype(np.int32)[0]
    correct_label = t
    pred_label = 1 if 0 < out else -1
    print(correct_label,pred_label,out,sep='\t')
    list_77.append([correct_label,pred_label,out])
    
```

## 77. 正解率の計測
76の出力を受け取り，予測の正解率，正例に関する適合率，再現率，F1スコアを求めるプログラムを作成せよ．

```python
# print(tuple(zip(*list_77)))
# (correct_label,pred_label,out) = tuple(zip(*list_77))
# print(np.array(correct_label) np.array(pred_label))

FN,FP,TN,TP = 0,0,0,0
for data in list_77:
    correct_label,pred_label,out = data
    FN += 1 if correct_label == 1 and pred_label == -1 else 0
    FP += 1 if correct_label == -1 and pred_label == 1 else 0
    TN += 1 if correct_label == -1 and pred_label == -1 else 0
    TP += 1 if correct_label == 1 and pred_label == 1 else 0

# correct =  np.sum((np.array(correct_label)*np.array(pred_label)) == 1)
# wrong = len(list_77) = correct
accuracy  = (TP+TN)/(FN+FP+TN+TP)
precision = (TP)/(FP+TP)
recall = (TP)/(FN+TP)
f1 = 2*precision*recall/(precision+recall)
print("accuracy : {:<3.2f}precision: {:<3.2f} recall:{:3.2f} f1: {:<3.2f}".format(accuracy,precision,recall,f1))
```

```python
def cal_data(data_list):
    FN,FP,TN,TP = 0,0,0,0
    for data in data_list:
        correct_label, pred_label,out = data
        FN += 1 if correct_label == 1 and pred_label == -1 else 0
        FP += 1 if correct_label == -1 and pred_label == 1 else 0
        TN += 1 if correct_label == -1 and pred_label == -1 else 0
        TP += 1 if correct_label == 1 and pred_label == 1 else 0
        
    accuracy  = (TP+TN)/(FN+FP+TN+TP) if FN+FP+TN+TP != 0 else 0
    precision = (TP)/(FP+TP) if FP+TP != 0 else 0
    recall = (TP)/(FN+TP) if FN+TP != 0 else 0
    f1 = 2*precision*recall/(precision+recall) if precision+recall != 0 else 0
    print("accuracy : {:<3.2f}precision: {:<3.2f} recall:{:3.2f} f1: {:<3.2f}".format(accuracy,precision,recall,f1))
```

```python
cal_data(list_77)
```

## 78. 5分割交差検定
76-77の実験では，学習に用いた事例を評価にも用いたため，正当な評価とは言えない．すなわち，分類器が訓練事例を丸暗記する際の性能を評価しており，モデルの汎化性能を測定していない．そこで，5分割交差検定により，極性分類の正解率，適合率，再現率，F1スコアを求めよ．

```python
def SplitData(data,n=3):
    
    ret = []
    data_size = len(data)
    split_size = [data_size//n for i in range(n)]
    for i in range(data_size%n):
        split_size[i] += 1

    index = 0
    for i in range(len(split_size)):
        ret.append(data[index:index+split_size[i]])
        index += split_size[i]
    return ret
```

```python
split_feat_data = SplitData(feat_list,n=5)
split_label_data = SplitData(label_list,n=5)

for i in range(len(split_feat_data)):
    train_data = []
    train_label = []
    for j in range(len(split_feat_data)):
        if i == j:
            test_data = split_feat_data[i]
            test_label = split_label_data[i]
        else:
            train_data += split_feat_data[i]
            train_label += split_label_data[i]
    
    batch_size = 64
    dataset = torch.utils.data.TensorDataset(torch.FloatTensor(train_data),torch.FloatTensor(train_label))
    data_loader = DataLoader(dataset,batch_size=batch_size,shuffle=True)
    
    model = Model()
    model.trainning(data_loader)
    data = model.cal(test_data,test_label)
    cal_data(data)
```
